```python
fig = plt.figure(figsize = (12, 12))
row = 3
col = 3

for i in range(1, 8, 3):
  dp_num = cv2.imread(dolphin_img_list[i])
  sk_num = cv2.imread(shark_img_list[i])
  wh_num = cv2.imread(whale_img_list[i])

  ax1 = fig.add_subplot(row, col, i)
  ax2 = fig.add_subplot(row, col, i+1)
  ax3 = fig.add_subplot(row, col, i+2)

  ax1.imshow(dp_num[:, :, ::-1])
  ax2.imshow(sk_num[:, :, ::-1])
  ax3.imshow(wh_num[:, :, ::-1])

  if i == 1:
    ax1.set_title('dolphin', fontsize = 18)
    ax2.set_title('shark', fontsize = 18)
    ax3.set_title('whale', fontsize = 18)

fig.suptitle('Sea animals', fontsize = 25)
plt.show()
```

```python
def make_box(file_num, seg):
  img = cv2.imread(file_num)

  height_leng = len(img[0])


  height_interval = height_leng // seg


  for _ in range(1, seg):
    img[height_interval, 0] = (0, 0, 0)
    height_interval += height_interval

  plt.imshow(img[:, :, ::-1])

make_box(shark_img_list[10], seg = 5)
```

```python
img = cv2.imread(shark_img_list[10])
seg = 5

height_leng = len(img[:,])
weight_leng = len(img[0,:])

height_num = 20
height_interval = height_leng // seg

for i in range(1, height_leng+1, height_interval):
  if 0 < i <= int(height_leng * 0.2):
    img[height_num : height_num+7, 8:weight_leng-8] = (0, 0, 0)
    height_num += height_interval

  elif int(height_leng * 0.2) < i <= int(height_leng * 0.4):
    img[height_num : height_num+7, 8:weight_leng-8] = (255, 0, 0)
    height_num += height_interval

  elif int(height_leng * 0.4) < i <= int(height_leng * 0.6):
    img[height_num : height_num+7, 8:weight_leng-8] = (0, 255, 0)
    height_num += height_interval

  elif int(height_leng * 0.6) < i <= int(height_leng * 0.8):
    img[height_num : height_num+7, 8:weight_leng-8] = (0, 0, 255)
    height_num += height_interval

  else:
    img[height_num : height_num+7, 8:weight_leng-8] = (255, 255, 255)
cv2_imshow(img)
```
